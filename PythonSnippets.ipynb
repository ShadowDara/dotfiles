## Print Message with 'abcd...'

This function will print a Message, for example `Hallo` and will start with `a` then `b` then `c` until it reached the character `H` then `Ha`, `Haa`, `Hab`... !

**Install**
```sh
pip install time
```

````python
import time

__alphabet__ = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Defines the pause between single characters
__text_sleep__ = 0.01

def crazy_print(msg):
    word = ""
    for char in msg:
        for _char_ in __alphabet__:
            print(f"{word}{_char_}", flush=True)
            time.sleep(__text_sleep__)
            if char == _char_:
                word += _char_
                break

            if _char_ == __alphabet__[-1] and char not in __alphabet__:
                print(f"{word}{char}", flush=True)
                time.sleep(__text_sleep__)
                word += char

crazy_print("Lol")
````

## Print Message with 'abcd...' with Rich Support

*Same as without `rich` but now with colors!*

**Sadly `rich` does not work in a Jupiter Notebook!**

**Install**
```sh
pip install time
pip install rich
```

````python
import time
from rich import print

__alphabet__ = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Defines the pause between single characters
__text_sleep__ = 0.01

def crazy_print_rich(color, msg):
    word = ""
    for char in msg:
        for _char_ in __alphabet__:
            print(f"{color}{word}{_char_}", flush=True)
            time.sleep(__text_sleep__)
            if char == _char_:
                word += _char_
                break

            if _char_ == __alphabet__[-1] and char not in __alphabet__:
                print(f"{color}{word}{char}", flush=True)
                time.sleep(__text_sleep__)
                word += char

crazy_print_rich("[bold green]", "Lol")
````
